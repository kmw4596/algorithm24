예제 4.5
=======

#### -이진 탐색(순환구조)-

### 1.1 문제 정의

- 순환 구조를 이용한 이진탐색을 활용하여 리스트를 반으로 나누어서 찾고자 하는 항목 범위를 좁혀가며 찾아가는 코드를 만들어라.

### 1.2 알고리즘 설명
- 이진 탐색 알고리즘은 정렬된 리스트에서 원하는 값을 찾는 빠른 방법입니다. 리스트를 반으로 나누어 찾고자 하는 값과 비교하여 탐색 범위를 절반으로 줄입니다. 이 과정을 반복하여 원하는 값을 찾거나 탐색 범위가 없어질 때까지 진행합니다. 이 알고리즘은 리스트의 크기에 비례하지 않고 로그 시간 안에 값을 찾을 수 있어 효율적입니다.

### 1.4 알고리즘 개요

1. 이진 탐색은 정렬된 리스트를 이용하여 특정 값을 찾는 알고리즘으로, 리스트를 반으로 나누어 탐색 범위를 좁혀가는 방식입니다.
2. 탐색 범위가 좁혀질 때까지 반복적으로 중간 항목을 선택하여 찾고자 하는 값과 비교합니다.
3. 찾고자 하는 값이 중간 항목보다 작으면 왼쪽 부분을, 크면 오른쪽 부분을 탐색합니다.
4. 이러한 과정을 반복하여 원하는 값을 찾거나 탐색 범위가 더 이상 없을 때까지 진행합니다.

### 1.5 알고리즘 코드

```python
def binary_search(A, key, low, high):
    if low <= high:
        mid = (low + high) // 2
        if key == A[mid]:
            return mid
        elif key < A[mid]:
            return binary_search(A, key, low, mid - 1)
        else:
            return binary_search(A, key, mid + 1, high)
    return -1
```

### 1.6 테스트 코드

```python
test_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
key = 5
result = binary_search(test_list, key, 0, len(test_list) - 1)
if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```
