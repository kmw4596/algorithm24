=======

#### -거듭제곱(억지 기법)-

### 1.1 문제 정의

- 억지기법을 활용하여 실수 x의 정수 n 거듭제곱 값을 구하는 코드를 만들어라

### 1.2 알고리즘 설명
- 거듭 제곱(억지 기법)은 단순한 반복문을 사용하여 주어진 실수 x의 정수 n 거듭제곱 값을 계산합니다. 반복문을 통해 x를 n번 곱하는 방식으로 동작합니다.

### 1.3 손으로 푼 예제


### 1.4 알고리즘 개요

1. 주어진 실수 x의 정수 n 거듭제곱 값을 계산하는 방법입니다.
2. 반복문을 사용하여 x를 n번 곱하는 방식으로 작동합니다.
3. 초기값으로 1.0을 설정하고, 반복문을 통해 n번만큼 x와 곱해나갑니다.
4. 최종 결과를 반환합니다.

### 1.5 알고리즘 코드

```python
def slow_power(x, n):
    result = 1.0
    for i in range(n):
        result = result * x
    return result
```

### 1.6 테스트 코드

```python
x = 2
n = 3
result = slow_power(x, n)
print(f"{x}의 {n} 거듭제곱 값은 {result}입니다.")
```

### 1.8 복잡도 분석
- 이 알고리즘은 단순한 반복문을 사용하여 주어진 x의 n 거듭제곱 값을 계산하므로 시간 복잡도는 O(n)입니다.
