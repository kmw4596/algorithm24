#### -삽입 정렬-

### 1.1 문제 정의

- 주어진 리스트 안의 요소들을 작은 것부터 큰 것까지 순서대로 재배치하는 코드를 삽입 정렬을 활용하여 만들어라.

### 1.2 알고리즘 설명
- 삽입 정렬 알고리즘은 리스트를 두 부분으로 나누어 생각합니다. 이미 정렬된 부분과 정렬되지 않은 부분입니다. 정렬되지 않은 부분의 원소를 정렬된 부분에 적절한 위치에 삽입하여 전체 리스트가 정렬되도록 합니다.

### 1.4 알고리즘 개요

1. 주어진 리스트를 가져옵니다.

2. 리스트를 이미 정렬된 부분과 정렬되지 않은 부분으로 나누어 생각합니다.
초기에는 첫 번째 요소만이 이미 정렬된 부분에 속합니다. 나머지 요소들은 정렬되지 않은 부분에 속합니다.

3. 정렬되지 않은 부분에서 하나의 요소를 선택하여 이미 정렬된 부분에 삽입하는 과정을 반복합니다.
선택된 요소가 이미 정렬된 부분의 적절한 위치를 찾을 때까지 왼쪽으로 이동합니다.

4. 선택된 요소를 적절한 위치에 삽입합니다.

5. 정렬되지 않은 부분에 요소가 남아있지 않을 때까지 이 과정을 반복합니다.

6. 모든 요소가 정렬된 상태로 정렬이 완료됩니다.

### 1.5 알고리즘 코드

```python
def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = key
        printStep(A, i)
```

```python
test_list = [5, 2, 4, 6, 1, 3]
insertion_sort(test_list)
print("Sorted list:", test_list)
```
