예제 4.11
=======

#### -축소 정복 기법을 이용한 k번째 작은 수 찾기-

### 1.1 문제 정의

- 축소 정복 기법을 활용하여 k번째 작은수를 찾는 코드를 만들어라.

### 1.2 알고리즘 설명
- 이 알고리즘은 퀵 정렬의 아이디어를 응용하여 동작합니다. 리스트를 분할하고 분할된 부분에서 pivot을 선택하여 리스트를 분할하는 과정을 반복합니다. 이 때, pivot의 위치가 k-1일 때 해당 값은 k번째로 작은 값이 됩니다.

### 1.3 손으로 푼 예제


### 1.4 알고리즘 개요

1. 주어진 리스트를 분할하고 pivot을 선택합니다.
2. pivot의 위치가 k-1이 될 때까지 리스트를 분할하고 pivot을 선택하는 과정을 반복합니다.
3. pivot의 위치가 k-1이면 해당 값을 반환합니다.

### 1.5 알고리즘 코드

```python
def quick_select(A, left, right, k):
    pos = partition(A, left, right)
    if pos + 1 == left + k:
        return A[pos]
    elif pos + 1 > left + k:
        return quick_select(A, left, pos - 1, k)
    else:
        return quick_select(A, pos + 1, right, k - (pos - 1 - left))
```

### 1.6 테스트 코드

```python
test_list = [3, 2, 1, 5, 4]
k = 3
result = quick_select(test_list, 0, len(test_list) - 1, k)
print(f"{k}번째 작은 수는 {result}입니다.")
```

### 1.8 복잡도 분석
- 이 알고리즘은 퀵 정렬의 partition 과정을 사용하므로 시간 복잡도는 최악의 경우 O(n^2), 평균적으로 O(n)입니다.
